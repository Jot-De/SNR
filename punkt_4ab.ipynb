```python
import pandas as pd
import sklearn.metrics as skm
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score


import itertools

# Create a confusion matrix
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title, fontsize=14)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.4f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",fontsize=18,
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')

"""
X_train = pd.read_csv('SVM_4a_X_train.csv').values
y_train = pd.read_csv('SVM_4a_y_train.csv').values

X_test = pd.read_csv('SVM_4a_X_test.csv').values
y_test = pd.read_csv('SVM_4a_y_test.csv').values
"""
"""
X_train = pd.read_csv('SVM_4b_X_train.csv').values
y_train = pd.read_csv('SVM_4b_y_train.csv').values

X_test = pd.read_csv('SVM_4b_X_test.csv').values
y_test = pd.read_csv('SVM_4b_y_test.csv').values

"""

#print(y_train.shape)
#print(X_train.shape)



C=[1]

for c in C:
    svm_model_linear = SVC(kernel='linear', C=c, probability=True)
    svm_model_quad = SVC(kernel='poly', C=c, probability=True, degree=2)
    svm_model_expo = SVC(kernel='rbf', C=c, probability=True)
    SVC = [svm_model_linear, svm_model_quad, svm_model_expo]
    for svc in SVC:
        clf = svc
        clf.fit(X_train, y_train)
        y_pred_svc=clf.predict(X_test)
        labels=y_train.uniqe()
        svc_cf = confusion_matrix(y_test, y_pred_svc)
        plot_confusion_matrix(svc_cf, labels, title=str(svc)+" C="+str(c)+" \n Confusion Matrix", cmap=plt.cm.Oranges, normalize=True)
        plt.show()
        from sklearn.metrics import classification_report
        print(classification_report(y_test, y_pred_svc))
        pred_svc=clf.decision_function(X_test)
        svc_fpr, svc_tpr, svc_threshold = roc_curve(y_test, pred_svc)
        print('Support Vector Classifier: ', roc_auc_score(y_train, pred_svc))
        plt.plot(svc_fpr, svc_tpr, label='Support Vector Classifier Score: {:.4f}'.format(roc_auc_score(y_train, pred_svc)))
        plt.plot([0, 1], [0, 1], 'k--')
        plt.axis([-0.01, 1.01, 0, 1.02])
        plt.show()
```
